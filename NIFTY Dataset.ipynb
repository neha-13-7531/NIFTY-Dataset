# Seasonality
- Understanding the patterns and deriving insights grgarding the repeated patterns.

# Types of Seasonality

**1.Additive Seasonality**
- The patterns are constant over time.
- Seasonal effect does not depend on the level of data.
- Shape of patterns remain same even if overall values are increase or decrease.
***y(t)= trend + seasonality + residuals***

**2.Multiplicative Seasonality**
- Patterns grows or shrinks depending on the level of the data.
- The seasonal effect multiplies with trend.
***y(t)= trend * seasonality * residuals***

**3.Time Based Seasonality**
- It depends on time period over which seasonality occurs.
- It can be daily seasonality,weekly seasonality,year seasonality,monthly seasonality.

**4.Short term-long term seasonality**
- In short term seasonality cycles within shorter period of time like days or weeks.
- In long termthe seasonality cycles in longer time period like months or years.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

period=30
# Additive Seasonality
add_dec = seasonal_decompose(df['Close'],model = 'additive',period = period)

components={
    'Observed':add_dec.observed,
    'Trend':add_dec.trend,
    'Seasonal':add_dec.seasonal,
    'Residual':add_dec.resid
}

for name,series in components.items():
    plt.figure(figsize=(8,4))
    plt.plot(series)
    plt.title(f"Additive Decomposition {name}")
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.show()
```

```python
# Multiplicative Seasonality
mul_dec = seasonal_decompose(df['Close'],model = 'multiplicative',period = period)

components={
    'Observed':mul_dec.observed,
    'Trend':mul_dec.trend,
    'Seasonal':mul_dec.seasonal,
    'Residual':mul_dec.resid
}

for name,series in components.items():
    plt.figure(figsize=(8,4))
    plt.plot(series)
    plt.title(f"Multiplicative Decomposition {name}")
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.show()
```
